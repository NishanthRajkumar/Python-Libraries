**Import required libraries**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
````

1. **Write a program to draw a scatter plot for random 1000 x and y coordinates**

````python
# plot scatter for random 1000 coordinates
x_val = np.random.randint(0, 10000, size=1000)
y_val = np.random.randint(0, 10000, size=1000)

plot = px.scatter(x=x_val, y=y_val, title='scatter for random 1000 coordinates')
plot.show()
````

2. **Write a program to draw line and scatter plots for random 100 x and y coordinates**

````python
# plot scatter for random 100 coordinates
x_val = np.random.randint(0, 1000, size=100)
y_val = np.random.randint(0, 1000, size=100)

plot = px.scatter(x=x_val, y=y_val, title='scatter for random 100 coordinates')
plot.show()
````

3. **Write a program to draw a scatter plot for random 500 x and y coordinates and style it**

````python
# plot scatter for random 500 coordinates with styling
x_val = np.random.randint(0, 1000, size=500)
y_val = np.random.randint(0, 1000, size=500)

plot = px.scatter(x=x_val, y=y_val, title='scatter for random 500 coordinates')
plot.update_layout(
    paper_bgcolor="#f5f5f1",
    plot_bgcolor="#b9b9cd"
)
plot.show()
````

4. **Write a program to draw a scatter plot for a given dataset and show datalabels on hover**

    https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

````python
# Load the dataset
usa_states_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv')
usa_states_data.head(5)
````

````python
# Scatter plot for the usa states dataset
plot = px.scatter(usa_states_data, x='State', y='Population', title='USA state wise population')
plot.update_traces(mode='markers')
plot.show()
````
