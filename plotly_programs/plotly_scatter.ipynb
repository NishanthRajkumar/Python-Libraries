**Import required libraries**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
````

1. **Write a program to draw a scatter plot for random 1000 x and y coordinates**

````python
# plot scatter for random 1000 coordinates
x_val = np.random.randint(0, 10000, size=1000)
y_val = np.random.randint(0, 10000, size=1000)

plot = px.scatter(x=x_val, y=y_val, title='scatter for random 1000 coordinates')
plot.show()
````

2. **Write a program to draw line and scatter plots for random 100 x and y coordinates**

````python
# plot scatter for random 100 coordinates
x_val = np.random.randint(0, 1000, size=100)
y_val = np.random.randint(0, 1000, size=100)

plot = px.scatter(x=x_val, y=y_val, title='scatter for random 100 coordinates')
plot.show()
````
