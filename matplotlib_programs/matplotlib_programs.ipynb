1. **Write a Python program to draw a line with suitable label in the x axis, y axis and a title**

````python
# Using pyplot, plot a simple straight line with appropriate label names
plt.axline((0,0), (5,6))
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('A straight line through (5,6) from origin')
plt.show()
````

2. **Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis and a title**

````python
# Plot a line given the axis values
def plot_line(xaxis_values, yaxis_values):
    """
        Description:
            Plot a line using the given axis values

        Parameter:
            xaxis_values: Values for x axis
            yaxis_values: Values for y axis
        
        Return:
            None
    """
    plt.plot(xaxis_values, yaxis_values)
    plt.title('Line for the given axis values')
    plt.xlabel('X axis values')
    plt.ylabel('Y axis values')
    plt.show()
````

````python
# Plot a line using the plot_line() function
plot_line([3,5,7,9], [2,5,6,8])
````

4. **Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016**

    Sample Financial data (fdata.csv)

````python
# read csv file using pandas module
financial_data = pd.read_csv('fdata.csv')
x_val = financial_data['Date']
y_open_val = financial_data['Open']
y_high_val = financial_data['High']
y_low_val = financial_data['Low']
y_close_val = financial_data['Close']

# Plot 4 lines for the open, high, low, close
plt.plot(x_val, y_open_val, label='Open', marker='o')
plt.plot(x_val, y_high_val, label='High', marker='o')
plt.plot(x_val, y_low_val, label='Low', marker='o')
plt.plot(x_val, y_close_val, label='Close', marker='o')
plt.legend()
plt.title('Alphabet Inc. stock price from October 3, 2016 to October 7, 2016')
plt.xlabel('Date')
plt.ylabel('Share value')
plt.show()
````

9. **Write a Python program to display the current axis limits values and set new axis values**

13. **Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with linestyle='-', width=0.5 and color=blue**

14. **Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks**

````python
# read csv file using pandas module
financial_data = pd.read_csv('fdata.csv')
x_val = financial_data['Date']
y_close_val = financial_data['Close']

# Plot line for close show grid and turn off ticks
plt.plot(x_val, y_close_val, marker='o', color='r')
plt.minorticks_on()
plt.grid(True, linestyle='-', color = 'b', linewidth = 0.75, which='major')
plt.grid(True, linestyle='-', color = 'g', linewidth = 0.5, which='minor', alpha= 0.2)
plt.tick_params(left=False, bottom=False)
plt.title('Alphabet Inc. closing stock price from October 3, 2016 to October 7, 2016')
plt.xlabel('Date')
plt.ylabel('Stock price')
plt.show()
````

15. **Write a Python program to create multiple plots**

````python
# read csv file using pandas module
financial_data = pd.read_csv('fdata.csv')
x_val = ['3-oct', '4-oct', '5-oct', '6-oct', '7-oct']
y_open_val = financial_data['Open']
y_high_val = financial_data['High']
y_low_val = financial_data['Low']
y_close_val = financial_data['Close']

# Create 2x2 subplots and plot open, high, low, close in each of them
figure, axis = plt.subplots(2,2)
axis[0,0].plot(x_val, y_open_val, label='Open', marker='d')
axis[0,0].set_title('Open')
axis[0,1].plot(x_val, y_high_val, label='High', marker='d')
axis[0,1].set_title('High')
axis[1,0].plot(x_val, y_low_val, label='Low', marker='d')
axis[1,0].set_title('Low')
axis[1,1].plot(x_val, y_close_val, label='Close', marker='d')
axis[1,1].set_title('Close')
plt.tight_layout()
plt.show()
````
