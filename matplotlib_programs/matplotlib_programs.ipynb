**Import required libraries**

````python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
````

1. **Write a Python program to draw a line with suitable label in the x axis, y axis and a title**

````python
# Using pyplot, plot a simple straight line with appropriate label names
plt.axline((0,0), (5,6))
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('A straight line through (5,6) from origin')
plt.show()
````

2. **Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis and a title**

````python
# Plot a line given the axis values
def plot_line(xaxis_values, yaxis_values):
    """
        Description:
            Plot a line using the given axis values

        Parameter:
            xaxis_values: Values for x axis
            yaxis_values: Values for y axis
        
        Return:
            None
    """
    plt.plot(xaxis_values, yaxis_values)
    plt.title('Line for the given axis values')
    plt.xlabel('X axis values')
    plt.ylabel('Y axis values')
    plt.show()
````

````python
# Plot a line using the plot_line() function
plot_line([3,5,7,9], [2,5,6,8])
````

3. **Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title**

    Test Data:\
    test.txt\
    1 2\
    2 4\
    3 1

````python
# Load the values from text file using numpy module
xy_arr = np.loadtxt('test.txt', dtype='int')
x_axis_val = xy_arr[:,0]
y_axis_val = xy_arr[:,1]

# Using the plot_line() function plot the line for the given axis values
plot_line(x_axis_val, y_axis_val)
````

4. **Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016**

    Sample Financial data (fdata.csv)

````python
# read csv file using pandas module
financial_data = pd.read_csv('fdata.csv')
x_val = financial_data['Date']
y_open_val = financial_data['Open']
y_high_val = financial_data['High']
y_low_val = financial_data['Low']
y_close_val = financial_data['Close']

# Plot 4 lines for the open, high, low, close
plt.plot(x_val, y_open_val, label='Open')
plt.plot(x_val, y_high_val, label='High')
plt.plot(x_val, y_low_val, label='Low')
plt.plot(x_val, y_close_val, label='Close')
plt.legend()
plt.title('Alphabet Inc. stock price from October 3, 2016 to October 7, 2016')
plt.xlabel('Date')
plt.ylabel('Share value')
plt.show()
````
