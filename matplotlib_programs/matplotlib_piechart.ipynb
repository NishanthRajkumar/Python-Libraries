**Import required libraries**

````python
from matplotlib import pyplot as plt
import pandas as pd
````

1. **Write a Python programming to create a pie chart of the popularity of programming Languages**

    Sample data:\
    Programming languages: Java, Python, PHP, JavaScript, C#, C++\
    Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

````python
# create list of the given data
programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Plot pie chart
plt.pie(popularity, labels=programming_languages)
plt.show()
````

2. **Write a Python programming to create a pie chart with a title of the popularity of programming Languages**

````python
# Plot pie chart with title
plt.pie(popularity, labels=programming_languages, autopct='%1.1f%%', explode=[0,0.1, 0, 0, 0, 0], shadow=True, startangle=-110)
plt.title('Programming language popularity')
plt.show()
````

3. **Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file**

    Sample data:\
    medal.csv\
    country,gold_medal\
    United States,46\
    Great Britain,27\
    China,26\
    Russia,19\
    Germany,17

````python
# Read from file and store as dataframe
medals_df = pd.read_csv('medal.csv')

# Plot pie chart for the 5 countries
plt.pie(medals_df['gold_medal'], labels=medals_df['country'], wedgeprops={'edgecolor': 'black'})
plt.title('Gold medal achievements of top 5 countries')
plt.show()
````
