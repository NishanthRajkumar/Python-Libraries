**Import required libraries**

````python
from matplotlib import pyplot as plt
import random
import numpy as np
````

1. **Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other**

````python
# get a random distribution for x and y
x=random.sample(np.arange(1,100).tolist(), k=50)
y=random.sample(np.arange(1,100).tolist(), k=50)

# Plot the scatter plot
plt.scatter(x, y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter plot of random distribution x and y')
plt.show()
````

2. **Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other**

````python
# get a random distribution for x and y
x=random.sample(np.arange(1,100).tolist(), k=50)
y=random.sample(np.arange(1,100).tolist(), k=50)

# Plot the scatter plot with circles
plt.scatter(x, y, marker='o', facecolors= 'none', edgecolors='b', s=70)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter plot of random distribution x and y')
plt.show()
````

3. **Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes**

````python
# get a random distribution for x and y and the sizes for the balls
x=random.sample(np.arange(1,100).tolist(), k=50)
y=random.sample(np.arange(1,100).tolist(), k=50)
sizes=random.sample(np.arange(20,150).tolist(), k=50)

# Plot the scatter plot with different sizes
plt.scatter(x, y, s=sizes, edgecolors='b')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter plot of random distribution x and y')
plt.show()
````
